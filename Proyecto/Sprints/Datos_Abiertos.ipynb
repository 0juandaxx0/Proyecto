## Limpieza de datasets

```python
# Importar librerias necesarias para analisis
import numpy as np
import unicodedata
```

### Data 1 - Informe de pasajeros 

```python
# Ver informacion general para conocer columnas y que datos podemos usar
print(df1.info()) # ver los tipos de datos que tenemos
print(df1.columns) # Ciudad, Pasajeros día típico laboral, DíaSemana
print(df1["Ciudad"].unique()) # Ver que ciudades tiene el dataset
```

```python
# 1. guardar solo las columnas que nos sirves para el analisis 
data1 = df1[["Ciudad", "Pasajeros día típico laboral", "DíaSemana"]]

# 2. Filtrar la ciudad a solo Bogota y Soacha
data1 = data1[data1["Ciudad"] == "Bogotá y Soacha"].copy()
data1["Ciudad"] = data1["Ciudad"].astype("string") # Convertir a cadena de caracteres

# 3. Convertir los tipos de datos de dia tipico laboral
print("Valores nulos o sin marcar: ", data1["Pasajeros día típico laboral"].isna().sum())
# reemplazamos valores de espacio y que tengan antes 
data1["Pasajeros día típico laboral"] = (
    data1["Pasajeros día típico laboral"]
    .astype("string") # convertir primero a cadena para reemplazar 
    .str.replace(".", "", regex=False)# quitar puntos de miles
    .str.strip() # quitar espacios en blanco
)
# pasamos a tipo entero
data1["Pasajeros día típico laboral"] = pd.to_numeric(data1["Pasajeros día típico laboral"]).fillna(0).astype(int)

# 4. Filtrar y cambiar dias de semana (Lunes a Viernes: 1 a 5)
data1 = data1[data1["DíaSemana"].isin([1,2,3,4,5])]

mapa_dias = {1:"lunes", 2:"martes", 3:"miercoles", 4:"jueves", 5:"viernes"}
data1["Dia"] = data1["DíaSemana"].map(mapa_dias)

# 5. Funcion para quitar tildes de variables
def quitar_tildes(texto):
    if pd.isna(texto):
        return texto
    # Normaliza el texto para separar las letras de sus acentos y elimina los acentos
    texto_norm = unicodedata.normalize('NFD', str(texto))
    return "".join([c for c in texto_norm if unicodedata.category(c) != 'Mn'])

# llamar funcion y dejar los nombres de las columnas y las filas en minusculas y sin tildes:
data1.columns = [quitar_tildes(col).lower().replace(" ", "_") for col in data1.columns]
data1["ciudad"] = data1["ciudad"].astype("string").str.lower().apply(quitar_tildes).str.strip()

# 6. Guardar dataset en csv para analisis
data1.to_csv('../datos_limpios/pasajeros_dia.csv', index=False, encoding="utf-8")
```

### Data 2 - validaciones diarias SITP

```python
# Ver informacion general para conocer columnas y que datos podemos usar
print(df2["Estacion_Parada"].unique()) # (07503) SAN MATEO - C.C. UNISUR, (07504) TERREROS
print(df2.columns) # Acceso_Estacion, Estacion_Parada
```

```python
# 1. Usar las columnas necesarias 
data2 = df2[["Acceso_Estacion", "Estacion_Parada"]]
data2 = data2.copy() # copia para trabajar en otro data

# Convertir a cadenas de texto
data2['Acceso_Estacion'] = data2['Acceso_Estacion'].astype("string")
data2['Estacion_Parada'] = data2["Estacion_Parada"].astype("string")

# 2. Filtrar la Estacion_Parada para san mateo y terreros (lugares donde llega el transmilenio)
estaciones = ["(07503) SAN MATEO - C.C. UNISUR", "(07504) TERREROS", "(57503) Ampliacion San Mateo"]
data2 = data2[data2["Estacion_Parada"].isin(estaciones)].copy()

# 3. limpiar columnas, quitando numeros y dejando solo nombre 
# Limpiamos todos los nombre de Acceso_Estacion para solo ver el nombre de la estacion
data2['Acceso_Estacion'] = (
    data2['Acceso_Estacion']
    .str.replace(r'\(.*?\)', '', regex=True) # Borra los caracteres dentro de parentesis, etc.
    .str.replace(r'\d+', '', regex=True)     # Borra números sueltos si los hay
    .str.strip()                             # Quita espacios dobles o al inicio/final
)

# limpiamos el nombre de las dos estaciones paradas 
data2['Estacion_Parada'] = (
    data2['Estacion_Parada']
    .astype(str)
    .str.replace(r'\(.*?\)', '', regex=True) # Quita lo que esté entre paréntesis
    .str.strip()
)

# 4. Funcion para quitar tildes de variables
def quitar_tildes(texto):
    if pd.isna(texto):
        return texto
    # Normaliza el texto para separar las letras de sus acentos y elimina los acentos
    texto_norm = unicodedata.normalize('NFD', str(texto))
    return "".join([c for c in texto_norm if unicodedata.category(c) != 'Mn'])

# pasar a minusculas toda la data
# llamar funcion y dejar los nombres de las columnas y las filas en minusculas y sin tildes:
data2.columns = [quitar_tildes(col).lower().replace(" ", "_") for col in data2.columns]
data2["acceso_estacion"] = data2["acceso_estacion"].astype("string").str.lower().apply(quitar_tildes).str.strip()
data2["estacion_parada"] = data2["estacion_parada"].astype("string").str.lower().apply(quitar_tildes).str.strip()

# 5. Guardar dataset en csv para analisis
data2.to_csv('../datos_limpios/validacion_SITP.csv', index=False, encoding="utf-8")
```

### Data 3 - Tiempo de desplazamiento promedio

```python
# Ver informacion general para conocer columnas y que datos podemos usar
df3.head(20)
```

```python
# 1. generar una columna con la velocidad promedio de cada año
# formula -> velocidad = Distancia/Tiempo
data3 = df3.copy()

# convertir minutos a hora -> velocidad = 10/(minutos/600) = 600/minutos
# Saber si hay valores 0 o nulos, poner un valor de 0 en velocidad
data3["Velocidad km/h"] = np.where(
    data3["Minutos en 10 Kilómetros recorridos"] > 0,
    600 /data3["Minutos en 10 Kilómetros recorridos"],
    0
)

# cambiar el nombre de la columna Unnamed: 0 a Año
data3 = data3.rename(columns={"Unnamed: 0": "Año"})
print(data3.head())

# 2. Funcion para quitar tildes de variables
def quitar_tildes(texto):
    if pd.isna(texto):
        return texto
    # Normaliza el texto para separar las letras de sus acentos y elimina los acentos
    texto_norm = unicodedata.normalize('NFD', str(texto))
    return "".join([c for c in texto_norm if unicodedata.category(c) != 'Mn'])

# llamar funcion y dejar los nombres de las columnas en minúsculas y sin tildes:
data3.columns = [quitar_tildes(col).lower().replace(" ", "_") for col in data3.columns]

# 3. guardar dataset en un csv para analisis
data3.to_csv('../datos_limpios/Tiempo_promedio.csv', index=False, encoding="utf-8")
```
