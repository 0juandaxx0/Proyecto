## Limpieza de Dataset

```python
# importe de librerias necesarias
import unicodedata
```

```python
# 1. Eliminar columnas innecesarias almacenar en una lista y luego hacer drop
columnas_a_eliminar = [
    'Marca temporal', 
    'Nombre completo', 
    'Correo electrónico institucional '
]
df = df.drop(columns=columnas_a_eliminar)
```

```python
# 2. Funcion para limpiar texto: minusculas, sin tildes, sin espacios extra (se llama mas adelante)
# funcion para limpiar los textos, convertir a minusculas y quitar tildes
def limpiar_texto(texto):
    if pd.isna(texto):
        return texto
    texto = str(texto).lower().strip()
    # Normalizar para quitar tildes
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    return texto
```

```python
# 3. Renombrar las columnas para facilitar su manejo
nuevos_nombres = {
    'Programa académico al que pertenece': 'programa',
    'Semestre actual que esta cursando': 'grupo_semestre', # pasara a ser un grupo_semestre
    'Jornada recurrente de clases o asistencia a la universidad': 'jornada',
    'Lugar de residencia': 'residencia',
    'Principal medio de transporte utilizado para la universidad': 'medio_transporte',
    '¿Cuánto tiempo en promedio tarda en llegar a la universidad?': 'tiempo_llegada',
    '¿Cuánto tiempo en promedio tarda en regresar de la universidad a  su lugar de residencia? ': 'tiempo_regreso',
    'Realizas trasbordos en tu trayecto (cambios de transporte a lo largo del trayecto)': 'trasbordos',
    '¿En qué tipo de horario realiza la mayor parte del desplazamiento?': 'hora_desplazamiento',
    '¿El tiempo de desplazamiento es variado? (escala de 1 a 5)': 'tiempo_desplazamiento',
    '¿Cuántas veces a la semana tiene que asistir a clases presenciales?': 'dias_asistencia',
    '¿Cuántas veces llega tarde a clases en la semana debido al transporte?': 'llegadas_tarde',
    '¿Cuántas clases a la semana pierde por problemas de movilidad?': 'pierde_clase',
    '¿Cuál es tu promedio académico actual?': 'promedio_academico',
    '¿Considera qué el tiempo de desplazamiento afecta tu rendimiento y puntualidad?': 'afecta_rendimiento',
    '¿Cuántas horas al día dedica a estudiar y hacer tareas?': 'horas_estudiar',
    '¿Considera qué pierde tiempo para estudiar debido a la movilidad que realizar diariamente?': 'pierde_tiempo',
    '¿Qué actividad realiza durante los desplazamientos?': 'actividades_desplazamiento',
    '¿Aceptaría cambiar el horario si esto redujera el tiempo de desplazamientos?': 'cambia_horario'
}
df = df.rename(columns=nuevos_nombres)

# Aplicar limpieza de texto a todas las columnas de tipo objeto (cadenas)
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(limpiar_texto) # llamar la funcion para limpiar textos
```

```python
# 4. Transformaciones y agrupaciones mas especificas

# Grupo de semestres a valores numéricos (1 al 5) donde 1 es semestres 1 y 2 y 9 es semestre 9
mapeo_semestres = {
    '1, 2 semestre': 1,
    '3, 4 semestre': 2,
    '5, 6 semestre': 3,
    '7, 8 semestre': 4,
    '9 o superior': 5
}
df['grupo_semestre'] = df['grupo_semestre'].map(mapeo_semestres)

# Hora de desplazamiento
mapeo_desplazamiento = {
    'ambos': "ambos",
    'horas valle (poca congestion vehicular y de personas)': "valle",
    'horas pico (mucha congestion vehicular y de personas)': "pico" 
}
df['hora_desplazamiento'] = df['hora_desplazamiento'].map(mapeo_desplazamiento)

# Trasbordos
mapeo_trasbordos = {
    'ninguno': 0,
    '1': 1,
    '2': 2,
    'mas de 3': 3 # Agrupamos 3 o mas como 3 (numérico)
}
df['trasbordos'] = df['trasbordos'].map(mapeo_trasbordos)

# Tiempos de desplazamiento (Llegada y Regreso) -> Mantenidos como cadenas limpias para agrupar
mapeo_tiempos_grupos = {
    'menos de 30 minutos (media hora o menos)': '0-30 min',
    '30 a 60 minutos (media a una hora)': '30-60 min',
    '61 a 90 minutos (una hora a hora y media)': '61-90 min',
    '91 a 120 minutos (hora y media a dos horas)': '91-120 min',
    'mas de 2 horas': 'Mas de 120 min'
}
df['tiempo_llegada'] = df['tiempo_llegada'].map(mapeo_tiempos_grupos)
df['tiempo_regreso'] = df['tiempo_regreso'].map(mapeo_tiempos_grupos)

# Extraer el valor numerico (1 a 5) de la escala de variabilidad
df['tiempo_desplazamiento'] = df['tiempo_desplazamiento'].str.extract(r'(\d+)').astype(int)

# Días de asistencia a presencial (ej: '5 dias' -> 5)
df['dias_asistencia'] = df['dias_asistencia'].str.extract(r'(\d+)').astype(int)

# Llegadas tarde semanales -> Escala ordinal/numerica
mapeo_tardes = {
    'ninguna': 0,
    '1 o 2 veces': 1.5,
    '3 o 4 veces': 3.5,
    '5 o mas veces': 5
}
df['llegadas_tarde'] = df['llegadas_tarde'].map(mapeo_tardes)

# Clases perdidas semanales -> Escala ordinal/numerica
mapeo_pierde = {
    'no pierdo clases': 0,
    '1 clase': 1,
    '2 clases': 2,
    '3 clases': 3,
    '4 o mas clases': 4
}
df['pierde_clase'] = df['pierde_clase'].map(mapeo_pierde)

# Horas dedicadas a estudiar -> Escala numerica (punto medio en horas)
mapeo_estudio = {
    'menos de 2 horas': 1,
    '3 a 4 horas': 3.5,
    '5 o mas horas': 5.5
}
df['horas_estudiar'] = df['horas_estudiar'].map(mapeo_estudio)
```

```python
# 6. Forzar tipos de datos finales

# Agregamos 'tiempo_llegada' y 'tiempo_regreso' a las categoricas
columnas_categoricas = [
    'programa', 'jornada', 'residencia', 'medio_transporte', 
    'tiempo_llegada', 'tiempo_regreso', 'hora_desplazamiento', 
    'afecta_rendimiento', 'pierde_tiempo', 'actividades_desplazamiento', 
    'cambia_horario', 'promedio_academico'
]
df[columnas_categoricas] = df[columnas_categoricas].astype("string")

# Quitamos las de tiempo de la lista numerica
columnas_numericas = [
    'grupo_semestre', 'trasbordos', 'tiempo_desplazamiento', 'dias_asistencia', 
    'llegadas_tarde', 'pierde_clase', 'horas_estudiar'
]

for col in columnas_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Verificacion
print(df[['tiempo_llegada', 'tiempo_regreso']].head())
```
